```python
# pip install plotly
# pip install --upgrade nbformat

# Import Library
import os
import pandas as pd

from var_engine.var_study import VaRStudy
from var_engine.data import DATA_SOURCE, EXAMPLE_PATH
from var_engine import plot_VaR

os.listdir(DATA_SOURCE)
```

```python
my_study = VaRStudy(EXAMPLE_PATH)
my_result = my_study.compute(
    start_date="08/09/2017", end_date="23/04/2024", window=365, percentile=0.95
)
```

```python
fig = plot_VaR(my_result)
fig.show()
```

```python
# Create a sample dataframe with one value per month for 5 years and coherent color changes
data = {
    'date': pd.date_range(start='2018-01-01', periods=60, freq='M'),
    'VaR': [1000 + i * 10 for i in range(60)],
    'confidence': [i / 59 for i in range(60)],  # Progression from 0 to 1
}
df = pd.DataFrame(data)
df.set_index('date', inplace=True)
fig = plot_VaR(df)
fig.show()
```
